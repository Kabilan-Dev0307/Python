"""
## 1. Pandas (Data Analysis)

Pandas is the most popular Python library used for data analysis [2, 3]. A Pandas DataFrame is similar to a database table or a DBMS table [4].
"""

````python
# Import the Pandas library using the conventional alias 'PD' [2]
import pandas as PD

# Create a list of data [2]. Note: A sample list is used here based on the filtering examples in [3].
data = [9,18,27,36,45]
print(data)
````

````python
# Convert the list into a Pandas DataFrame [2, 4]
# The resulting DataFrame is named 'my_data_frame' for subsequent operations [5]
my_data_frame = PD.DataFrame(data)
print(my_data_frame)
````

````python
# Use the .loc subsetting command to select rows 2 and 3 (index 2 and 3) [5]
# Note: The source indicates '2:3' selects row 2 and 3, implying inclusive indexing [5] 
print("Subsetting rows 2 and 3:")
my_data_frame.loc[1:4]
````

````python
# Conditional subsetting: Selecting values greater than 32 from column 0 [3]
# Note: The column name is '0' because the DataFrame was created from a single list [3].
print("Conditional subsetting (values > 32):")
my_data_frame.loc[my_data_frame [0] > 32]
````

### üêº Pandas Operations

Pandas performs other operations like:

- **Summarization**‚Äì Quickly compute statistics about your data.
- **Sorting**‚Äì Arrange your data by column values.
- **Reshaping**‚Äì Transform the structure of your dataset.
- **Aggregation**‚Äì Combine and summarize groups of data.


````python
# Sample DataFrame
data = {
    'Department': ['IT', 'HR', 'Finance', 'IT', 'HR', 'Finance'],
    'Employee': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Salary': [70000, 50000, 65000, 72000, 52000, 64000],
    'Experience': [5, 3, 6, 7, 2, 8]
}

df = PD.DataFrame(data)
df
````

````python
# Get a quick statistical summary
df.describe()
````

````python
# Sort by Salary (descending)
df.sort_values(by='Salary', ascending=False)
````

````python
# Pivot the data to view average salary by department
df.pivot_table(values='Salary', index='Department', aggfunc='mean')
````

````python
# Group by Department and calculate average salary & experience
df.groupby('Department')[['Salary', 'Experience']].mean()
````

## 2. NumPy (Numerical Python)

NumPy stands for numerical Python [6]. It is an excellent package used to perform mathematical and statistical operations [6]. NumPy is very popular because it consumes less memory and is very fast and convenient to use [7].

````python
# Import NumPy using the conventional alias 'NP' [6]
import numpy as NP

# Define a sample list (values used here are illustrative; the mean example given in the source is 49) [6]
my_list = [9,18,27,36,45]

# Calculate the mean of the list [6]
mean_value = NP.mean(my_list)
print(f"Mean value: {mean_value}") # Note: The source mentions 49 is the mean for their list [6].

# Calculate the median of the list [6]
median_value = NP.median(my_list)
print(f"Median value: {median_value}")

# NumPy can also compute mode, standard deviation, and other computations [6].
````

"""
## 3. Matplotlib (Plotting/Visualization)

Matplotlib is a Python library for plotting [7]. This package is very useful for exploratory data analysis (EDA) [8]. Matplotlib offers a wide variety of visualizations and charts [8]. There is an advanced version called Seaborn, which is also very useful for data science [8].

````python
# Import the PI plot module from matplotlib and alias it as 'PLT' [7]
from matplotlib import pyplot as PLT

# Creating two lists for plotting X and Y data [7]
list_X = [15,12,19,34,67]
list_Y = [10,15,18,14,12]

# Generate a line plot [7]
PLT.plot(list_X, list_Y)
PLT.show() # Display the plot [10]
````

````python
# Generate a bar plot [10]
PLT.bar(list_X, list_Y)
PLT.show()

# Generate a histogram [10]
PLT.hist(list_X) # Histograms typically use a single list of data
PLT.show()

# Generate a scatter plot [10]
PLT.scatter(list_X, list_Y)
PLT.show()
````

### üé® Seaborn ‚Äî Statistical Data Visualization Library

**Seaborn** is a powerful Python library built on top of **Matplotlib** that makes it easy to create **beautiful, informative, and statistical visualizations**.

---

- Works directly with **Pandas DataFrames**.
- Includes **themes and color palettes** for better aesthetics.
- Simplifies creation of **complex plots** like:
  - **Distribution plots** (`sns.histplot`, `sns.kdeplot`)
  - **Categorical plots** (`sns.boxplot`, `sns.violinplot`, `sns.barplot`)
  - **Regression plots** (`sns.regplot`, `sns.lmplot`)
  - **Matrix plots** (`sns.heatmap`, `sns.clustermap`)

---

#### üß† Why Use Seaborn?
Seaborn is great for **exploratory data analysis (EDA)** because it helps you:
- Understand data distributions quickly.
- Compare variables easily.
- Reveal trends and correlations visually.

### üéØ Seaborn Examples
Below are some common plots you can make with Seaborn for quick and effective data visualization.


````python
import seaborn as sns

# Load built-in dataset
tips = sns.load_dataset("tips")

# Histogram with density curve
sns.histplot(tips["total_bill"], kde=True, color="skyblue")
PLT.title("Distribution of Total Bill")
PLT.show()
````

````python
sns.lmplot(x="total_bill", y="tip", data=tips, height=5, aspect=1.2)
PLT.title("Total Bill vs Tip Amount")
PLT.show()
````

````python
sns.boxplot(x="day", y="total_bill", data=tips, palette="Set2")
PLT.title("Total Bill Distribution by Day")
PLT.show()
````

````python
sns.barplot(x="day", y="tip", data=tips, estimator=sum, ci=None, palette="muted")
PLT.title("Total Tips Collected by Day")
PLT.show()
````

````python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
PLT.title("Correlation Heatmap")
PLT.show()
````

"""
## 6. Pickle (Model Deployment)

Pickle is a Python library mostly used for deployment purposes [12]. When a machine learning algorithm is trained, the resulting model needs to be saved as a pickle object so that it can be consumed by a third party [12]. This process saves the learned pattern entity to disk [12].
"""

````python
# Import pickle and alias it as 'PK' [12]
import pickle as PK
````

````python
# Assume 'model' is the machine learning entity which has learned the pattern [12]
# For demonstration purposes, we define a dummy object 'model'
model = {'algorithm': 'Trained_Model_Placeholder'}

# Define the filename with the standard pickle extension (.sav) [12]
file_name = 'model.sav'

# 1. Saving (Dumping) the model object to disk [12]
# 'wb' stands for write binary
with open(file_name, 'wb') as file:
    PK.dump(model, file)
print(f"Model saved to disk as {file_name}")

# 2. Loading the saved model object [12]
# 'rb' stands for read binary
with open(file_name, 'rb') as file:
    loaded_model = PK.load(file)
print("Model successfully loaded:")
print(loaded_model)
````
