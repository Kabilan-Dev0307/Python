```python
import matplotlib.pyplot as plt
import numpy as np

v1 = np.array([3, 0])
v2 = np.array([3, 2])
v3 = np.array([3, 2])

plt.arrow(0, 0, v1[0], v1[1], head_width=0.2, head_length=0.3, fc='blue', ec='blue', label='Vector v1')
plt.arrow(v1[0], v1[1], 0, v2[1], head_width=0.2, head_length=0.3, fc='orange', ec='orange', label='Vector v2')
plt.arrow(0, 0, v3[0], v3[1], head_width=0.2, head_length=0.3, fc='green', ec='green', label='Vector v3')

plt.xticks(np.arange(-3, 5))
plt.yticks(np.arange(-3, 5))
plt.axvline()
plt.legend()
plt.grid()
plt.show()
```
